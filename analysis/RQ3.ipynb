```python
import pandas as pd
```

```python
agent_assinged_df = pd.read_csv("/Users/elmiraonagh/Desktop/courses/6444/project/project_repo/SWE-Manger/Project/RQ3/assigned_agents_2.csv")
agent_assinged_df = agent_assinged_df[['instance_id', 'cluster', 'assigned_agent']]
agent_assinged_df["success_state"] = ""
agent_assinged_df.head()
```

```python
agent_assinged_df["assigned_agent"].value_counts()
```

```python
GT_df = pd.read_csv("/Users/elmiraonagh/Desktop/courses/6444/project/project_repo/SWE-Manger/Project/RQ3/SWE_bench_lite_with_resolution.csv")
GT_df.head()
```

```python
success_states = []

for idx, row in agent_assinged_df.iterrows():
    instance_id = row['instance_id']
    agent = row['assigned_agent']
    
    # Lookup in GT_df: get the row for this instance, then the column for the agent
    match_row = GT_df[GT_df['instance_id'] == instance_id]
    if not match_row.empty:
        success_states.append(match_row[agent].values[0])
    else:
        success_states.append(None)  # if instance not found

# Add success_state column
agent_assinged_df['success_state'] = success_states

# Optional: count occurrences of assigned agents
agent_counts = agent_assinged_df['assigned_agent'].value_counts()
```

```python
# Group by agent and calculate success rate
success_rate_per_agent = agent_assinged_df.groupby('assigned_agent')['success_state'].mean()

print(success_rate_per_agent)
```

```python
# Overall success rate
overall_success_rate = agent_assinged_df['success_state'].mean()

print("Overall success rate:", overall_success_rate)
```
